```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\filip\Downloads\Medicare_Hospital_Spending_by_Claim.csv")
df
```

```python
#Question 1

df = df.drop(['Measure Start Date', 'Measure End Date'], 1)
df
```

```python
# Question 2

newColumns = ['Hospital', 'Provider ID', 'State', 'Period', 'Claim Type', 'Avg Spending Hospital', 'Avg Spending State', 
              'Avg Spending Nation', 'Percent Spending Hospital', 'Percent Spending State', 'Percent Spending Nation']
df.rename(columns=dict(zip(df.columns[0:], newColumns)),inplace=True)
df
```

```python
# Question 3

df2 = df[df.State == 'TX']
df2
```

```python
# Question 4

df2['Avg Spending Hospital'] = df2['Avg Spending Hospital'].str.replace(r'\$', '').astype(float)
avgspend = df2.groupby(['Claim Type', 'Hospital'])['Avg Spending Hospital'].mean().reset_index()
avgspend
```

```python
# Question 5
df2['Percent Spending Hospital'] = df2['Percent Spending Hospital'].str.replace(r'\%', '').astype(float)
percentspend = df2.groupby(['Claim Type', 'Hospital']).agg({'Percent Spending Hospital': ['mean', 'max']})
percentspend
```
