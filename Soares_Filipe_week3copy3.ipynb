```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
#create a histogram that fits a KDE
randomplot = np.random.normal(size=500)
sns.distplot(randomplot);
```

```python
# remove density curve and add rug plot
sns.distplot(randomplot, kde=False, rug=True);
```

```python
#adjust the number of bins to reveal different features in the data
sns.distplot(randomplot, bins=10, kde=False, rug=True);
```

```python
# remove bins, focus on density with KDE
sns.distplot(randomplot, hist=False, rug=True);
```

```python
# making a KDE
x = np.random.normal(0, 1, size=30)
wave = 1.06 * x.std() * x.size ** (-1 / 5.)
sup = np.linspace(-4, 4, 200)

kerns = []
for x_i in x:

    kern = stats.norm(x_i, wave).pdf(sup)
    kerns.append(kern)
    plt.plot(sup, kern, color="g")

sns.rugplot(x, color=".1", linewidth=4);
```

```python
#compute the value of density at each point
dense = np.sum(kerns, axis=0)
dense /= integrate.trapz(dense, sup)
plt.plot(sup, dense);
```

```python
#plot density with kdeplot
sns.kdeplot(x, shade=True);
```

```python
#KDE plot while setting bandwidth parameters
sns.kdeplot(x)
sns.kdeplot(x, bw=.1, label="bw: 0.1")
sns.kdeplot(x, bw=1, label="bw: 1")
plt.legend();
```

```python
#use 'cut' to influence how curve is drawn
sns.kdeplot(x, shade=True, cut=0)
sns.rugplot(x);
```

```python
#use distplot to fit paramatic distribution to random dataset
randpl = np.random.gamma(4, size=400)
sns.distplot(randpl, kde=False, fit=stats.gamma);
```

```python
#bivariate distribution using jointplot
mean, cov = [0, 2], [(2, .75), (.75, 2)]
rdmul = np.random.multivariate_normal(mean, cov, 200)
dafr = pd.DataFrame(rdmul, columns=["bottom", "side"])
```

```python
#visualize bivariate distribution with a scatterplot
sns.jointplot(x="bottom", y="side", data=dafr);
```

```python
#Hexbin
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="r");
```

```python
#KDE as a style
sns.jointplot(x="bottom", y="side", data=dafr, kind="kde");
```

```python
#2D KDE
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(dafr.bottom, dafr.side, ax=ax)
sns.rugplot(dafr.bottom, color="g", ax=ax)
sns.rugplot(dafr.side, vertical=True, ax=ax);
```

```python
#increase contour levels
fax = plt.subplots(figsize=(7, 7))
cont = sns.cubehelix_palette(as_cmap=True, dark=0, light=2, reverse=True)
sns.kdeplot(dafr.bottom, dafr.side, cmap=cont, n_levels=60, shade=True);
```

```python
grid = sns.jointplot(x="bottom", y="side", data=dafr, kind="kde", color="m")
grid.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
grid.ax_joint.collections[0].set_alpha(0)
grid.set_axis_labels("$X$", "$Y$");
```

```python
#using pairplot for multiple pairwise bivariate distributions
eyeris = sns.load_dataset("iris")
sns.pairplot(eyeris);
```

```python
#using pairGrid
grma = sns.PairGrid(eyeris)
grma.map_diag(sns.kdeplot)
grma.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6);
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "regression")))
```

```python
cash = sns.load_dataset("tips")
```

```python
# Drawing linear regression models with functions
sns.regplot(x="total_bill", y="tip", data=cash);
```

```python
sns.lmplot(x="total_bill", y="tip", data=cash);
```

```python
sns.lmplot(x="size", y="tip", data=cash);
```

```python
sns.lmplot(x="size", y="tip", data=cash, x_jitter=.05);
```

```python
sns.lmplot(x="size", y="tip", data=cash, x_estimator=np.mean);
```

```python
#Anscombe's quartet
anscombe = sns.load_dataset("anscombe")
```

```python
#Good model
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 80});
```

```python
#Same dataset, different plot, no bueno
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           robust = True, ci=None, scatter_kws={"s": 80});
```

```python
#using lmplot to fit polynomial regression model to find simple nonlineal trends
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});
```

```python
#outliers can be a problem
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           ci=None, scatter_kws={"s": 80});
```

```python
#use robust regressions to mitigate the effects of outliers
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
         robust = True,  ci=None, scatter_kws={"s": 80});
```

```python
#linear regression technically works here but predictions are not plausible
cash["lotsacash"] = (cash.tip / cash.total_bill) > .15
sns.lmplot(x="total_bill", y="lotsacash", data=cash,
           y_jitter=.03);
```

```python
#use logistic regression to help this
sns.lmplot(x="total_bill", y="lotsacash", data=cash,
           logistic=True, y_jitter=.03);
```

```python
#lowess smoother
sns.lmplot(x="total_bill", y="tip", data=cash,
           lowess=True);
```

```python
#use residplot to check if simple regression model is appropriate
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
              scatter_kws={"s": 80});
```

```python
#no structure above, that was good, this one below, not so good. the points should be scattered
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
              scatter_kws={"s": 80});
```

```python
#use lmplot to show linear relationship for multiple variables using colour to distinguish them
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=cash);
```

```python
#you can use different markers as well as colour
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=cash,
           markers=["x", "+"], palette="Set2");
```

```python
# adding more variables by drawing more facets
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=cash);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=cash);
```

```python
#control the size of the graph
p, axis = plt.subplots(figsize=(4, 4))
sns.regplot(x="total_bill", y="tip", data=cash, ax=axis);
```

```python
#change size of each facet
sns.lmplot(x="total_bill", y="tip", col="day", data=cash,
           col_wrap=4, height=5);
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=cash,
           aspect=.75);
```

```python
#using jointplot and regplot to show the linear regression on each axis with "kind=reg"
sns.jointplot(x="total_bill", y="tip", data=cash, kind="reg");
```

```python
#use pairplot to show relationships between multiple variables in a dataset
sns.pairplot(cash, x_vars=["total_bill", "size"], y_vars=["tip"],
             height=6, aspect=.57, kind="reg");
```

```python
#use hue to condition another categorical variable
sns.pairplot(cash, x_vars=["total_bill", "size"], y_vars=["tip"],
             hue="smoker", height=4, aspect=.7, kind="reg");
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
```

```python
np.random.seed(sum(map(ord, "categorical")))
```

```python
bigboat = sns.load_dataset("titanic")
cash = sns.load_dataset("tips")
eyeris = sns.load_dataset("iris")
```

```python
#categorical stripplot
sns.stripplot(x="day", y="total_bill", data=cash);
```

```python
#use swarmplot to prevent overlapping points
sns.swarmplot(x="day", y="total_bill", data=cash);
```

```python
#add a nested categorical variable with "hue"
sns.swarmplot(x="day", y="total_bill", hue="sex", data=cash);
```

```python
#sort the plots, seaborn will use either numerical order or the way the categories appear in dataframe or a set default order
sns.swarmplot(x="size", y="total_bill", data=cash);
```

```python
#orientation can be forced or inferred, here it is inferred
sns.swarmplot(x="total_bill", y="day", hue="time", data=cash);
```

```python
#box plots
sns.boxplot(x="day", y="total_bill", hue="time", data=cash);
```
